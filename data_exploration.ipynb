```python
import matplotlib.pyplot as plt

data = df.groupby('education')['monthly_salary'].mean().sort_values()

plt.figure(figsize=(8,5))
plt.bar(data.index, data.values)

plt.xlabel("Education Level")
plt.ylabel("Average Monthly Salary")
plt.title("Average Monthly Salary by Education")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average monthly salary wrt to age

data = df.groupby('age')['monthly_salary'].mean()
plt.title("Average Monthly Salary (Age)")
sns.lineplot(data)
```
